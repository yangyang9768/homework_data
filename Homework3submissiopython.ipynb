```python

import datetime
import pandas_datareader as pdr
import pandas as pd

api_key='56ee2bbb0aaca1ecbcee73bf32aa2f5e98d782cb'


tickers = ['BNTX', 'PFE', 'JNJ', 'MRNA']

#BNTX BIONTEX  PFE pfizer   JNJ J&J; MRNA: MODANA

start="2020-01-01"
end="2021-01-01"

df = pdr.tiingo.TiingoDailyReader(tickers, start=start, end=end, api_key=api_key)


all_data = df.read()
all_data
```

```python
# Import matplotlib
import matplotlib.pyplot as plt 

# Isolate the `Adj Close` values and transform the DataFrame
daily_close_px = all_data[['adjClose']].reset_index().pivot('date', 'symbol', 'adjClose')

# Calculate the daily percentage change for `daily_close_px`
daily_pct_change2021 = daily_close_px.pct_change()

# Plot the distributions
daily_pct_change2021.hist(bins=20, sharex=True, figsize=(12,8))

# Show the resulting plot
plt.show()
```
