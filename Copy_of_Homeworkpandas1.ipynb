Using any finance API (we use Yahoo Finance and Tiingo during sessions) download historical
stock prices for Google (symbol ‘GOOG’), for the period 1 Jan 2020 to 1 Jan 2021.
2. Save these prices to csv file with Pandas
3. Read this csv file to display data saved.
4. Plot a simple diagram to display closing prices for GOOG stock. 


```python
import pandas as pd
import pandas_datareader as pdr
# acquire_data_function
symbol_g='goog'
api_key_changxu='56ee2bbb0aaca1ecbcee73bf32aa2f5e98d782cb'

start="2020-1-1"
end="2021-1-1"

def acquire_data (symbol,api_key,start_date,end_data):
    df = pdr.tiingo.TiingoDailyReader(symbol, start=start, end=end, api_key=api_key)
    stock_price = df.read()
    return stock_price

acquire_data (symbol_g,api_key_changxu,start,end)

stock_price_g=acquire_data (symbol_g,api_key_changxu,start,end)

```

```python
# save data to csv

def save_data(source_data,file_name):
      source_data.to_csv(file_name,  date_format='%Y-%m-%d')
      saved_df = pd.read_csv(file_name, header=0, index_col='date', parse_dates=True)
      return saved_df



save_data(stock_price_g,'g_historical.csv')
```

```python
import matplotlib.pyplot as plt
```

```python
#Creat Plot

def create_diagram_close_plot(source_data):
   source=pd.read_csv(source_data)
   source["close"].plot(grid=True)
   plt.show()


create_diagram_close_plot('g_historical.csv')

```
